## Plot for simple_test

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

maddpg = pd.read_csv('./simple_test_maddpg.csv').iloc[:,5].to_numpy()
pretrain = pd.read_csv('./simple_test_load_pretrain.csv').iloc[:,5].to_numpy()
load_meta = pd.read_csv('./simple_test_load_meta.csv').iloc[:,5].to_numpy()
load_meta_10 = pd.read_csv('./simple_test_load_meta_rate=10.csv').iloc[:,5].to_numpy()
```

```python
def smooth(scalars, weight=0.6):
    last = scalars[0]
    smoothed = list()
    for point in scalars:
        smoothed_val = last * weight + (1 - weight) * point
        smoothed.append(smoothed_val)
        last = smoothed_val
    return smoothed
```

```python

with plt.style.context('seaborn-darkgrid'):
    plt.plot(range(len(load_meta)), smooth(load_meta), label='Meta-Q', color='#1874CD')
    plt.plot(range(len(maddpg)), smooth(maddpg), label='MADDPG', color='#EE8262')
    # plt.plot(range(len(pretrain)), smooth(pretrain), label='pretrain', color='#7AC5CD')
    
    # plt.plot(range(len(load_meta_10)), smooth(load_meta_10), label='load_meta_rate=10', color='#FF4040')
    plt.legend()
    plt.title('Simple Test', fontsize=13)
    plt.xlabel('episode', fontsize=12)
    plt.ylabel('rewards', fontsize=12)

plt.show()
```
